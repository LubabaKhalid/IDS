```python
import pandas as pd

# Load the Iris dataset
iris=pd.read_csv('/kaggle/input/iris.csv/other/default/1/Iris.csv')

# Check for missing values in the dataset
missing_data=iris.isnull().sum()
missing_data
```

```python
# List of numerical columns
numerical_columns=['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']

# Fill missing values in numerical columns with the median
for col in numerical_columns:
    iris[col] = iris[col].fillna(iris[col].median())  # Assign the filled column back to the DataFrame
```

```python
# Check and handle missing values in the 'Species' column
if iris['Species'].isnull().sum() > 0:
    iris['Species'] = iris['Species'].fillna(iris['Species'].mode()[0])  # Use the mode and assign it back
```

```python
#step2_1
# Remove duplicate rows in the dataset
iris.drop_duplicates(inplace=True)

# Verify the result by checking the number of rows before and after removing duplicates
print(f"Number of rows after removing duplicates: {iris.shape[0]}")
```

```python
#step2_2
# Create separate columns for sepal_area and petal_area
iris['sepal_area'] = iris['SepalLengthCm'] * iris['SepalWidthCm']
iris['petal_area'] = iris['PetalLengthCm'] * iris['PetalWidthCm']

# Create the total_area column by adding the sepal_area and petal_area
iris['total_area'] = iris['sepal_area'] + iris['petal_area']

# Display the updated dataset to check the new columns
print(iris[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm', 'sepal_area', 'petal_area', 'total_area']].head())
```

```python
#step2_3
# Drop rows that still have missing values after imputation
iris.dropna(inplace=True)

# Verify that there are no more missing values
missing_data_after_dropping = iris.isnull().sum()
print(missing_data_after_dropping)
```

```python
#step3_1
# Map species to numerical values
species_mapping = {'setosa': 0, 'versicolor': 1, 'virginica': 2}
iris['Species'] = iris['Species'].map(species_mapping)

# Verify the conversion
print(iris[['Species']].head())
```

```python
#step3_2
# Group the data by the transformed 'Species' column and calculate the sum of the numeric columns
aggregation_result = iris.groupby('Species')[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']].sum()

# Display the aggregation result
print(aggregation_result)
```

```python
#step4
# Reshape the dataset into a long format using the melt() function
iris_long = iris.melt(id_vars=['Species'], 
                     value_vars=['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm'],
                     var_name='Attribute', 
                     value_name='Value')

# Display the reshaped dataset
print(iris_long.head())
```

```python
df=pd.read_csv("/kaggle/input/employee-performance-and-salary-dataset/Employe_Performance_dataset.csv")
print(df.head(10))
```

```python
#step2
print(df.dtypes)
```

```python
#step3
# Get the number of rows and columns
rows, columns = df.shape

# Display the result
print(f"The dataset has {rows} rows and {columns} columns.")
```

```python
#step4
# Select the Name, Age, and Salary columns
selected_columns = df[['Name', 'Age', 'Salary']]

# Display the result
print(selected_columns)
```

```python
#step5
# Select the first 5 rows using iloc
first_5_rows = df.iloc[:5]

# Display the result
print(first_5_rows)
```

```python
#step6
# Select all rows where the Department is "Sales" using loc
sales_department = df.loc[df['Department'] == 'Sales']

# Display the result
print(sales_department)
```

```python
#step7
# Select rows where the Salary is greater than 8000 and Status is "Active"
filtered_rows = df.loc[(df['Salary'] > 8000) & (df['Status'] == 'Active')]

# Display the result
print(filtered_rows)
```

```python
#step8
# Extract rows where the employee has been with the company for more than 5 years
# and has a Performance Score of at least 4
filtered_employees = df.loc[(df['Experience'] > 5) & (df['Performance Score'] >= 4)]

# Display the result
print(filtered_employees)
```

```python
#step9
# Select every alternate row using iloc
alternate_rows = df.iloc[::2]

# Display the result
print(alternate_rows)
```

```python
#step10
# Calculate the threshold for the top 10% salary
salary_threshold = df['Salary'].quantile(0.9)

# Filter rows for "Sales" department and top 10% salary
sales_top_10_salary = df.loc[(df['Department'] == 'Sales') & (df['Salary'] >= salary_threshold)]

# Display the result
print(sales_top_10_salary)
```

```python
#step11
# Replace "Inactive" with "Retired" in the Status column
df['Status'] = df['Status'].replace('Inactive', 'Retired')

# Display the result
print(df['Status'].value_counts())  # Check the updated values in the Status column
```

```python
#step12
# Add a new column "Bonus" as 10% of the Salary
df['Bonus'] = df['Salary'] * 0.10

# Display the result
print(df[['Salary', 'Bonus']].head())
```

```python
#step13
# Update Performance Score to 5 for employees in IT department if the score is NaN
df.loc[(df['Department'] == 'IT') & (df['Performance Score'].isna()), 'Performance Score'] = 5

# Display the result
print(df[df['Department'] == 'IT'][['Name', 'Performance Score']].head())
```

```python
#step14
# Modify the Experience to 2 for employees with less than 2 years
df.loc[df['Experience'] < 2, 'Experience'] = 2

# Display the result
print(df[['Name', 'Experience']].head())
```

```python
#step15
# Normalize the Salary column between 0 and 1
df['Normalized Salary'] = (df['Salary'] - df['Salary'].min()) / (df['Salary'].max() - df['Salary'].min())

# Display the result
print(df[['Salary', 'Normalized Salary']].head())
```

```python
#step16
# Replace Location with "Remote" for employees with Performance Score below 3
df.loc[df['Performance Score'] < 3, 'Location'] = 'Remote'

# Display the result
print(df[['Name', 'Performance Score', 'Location']].head())
```

```python
#step17
# Create the Experience Level column based on Experience
conditions = [
    (df['Experience'] < 3),
    (df['Experience'] >= 3) & (df['Experience'] < 6),
    (df['Experience'] >= 6)
]

# Experience Level categories
experience_levels = ['Junior', 'Mid', 'Senior']

# Assign the Experience Level using np.select
import numpy as np
df['Experience Level'] = np.select(conditions, experience_levels, default='Unknown')

# Display the result
print(df[['Name', 'Experience', 'Experience Level']].head())
```

```python
# Define the new row as a DataFrame
new_row = pd.DataFrame({
    'ID': [9999],
    'Name': ['John Doe'],
    'Age': [30],
    'Gender': ['Male'],
    'Department': ['HR'],
    'Salary': [7500],
    'Joining Date': ['2024-01-01'],
    'Performance Score': [4],
    'Experience': [5],
    'Status': ['Active'],
    'Location': ['New York'],
    'Session': ['2024-Q1']
})

# Append the new row to the DataFrame using pd.concat
df = pd.concat([df, new_row], ignore_index=True)

# Display the updated DataFrame
print(df.tail())
```

```python
#step19
# Delete the 'Session' column
df = df.drop(columns=['Session'])

# Display the result
print(df.head())
```

```python
#step20
# Delete rows where Age is greater than 60
df = df[df['Age'] <= 60]

# Display the result
print(df.head())
```

```python
#step21
# Define conditions for Age Group
conditions = [
    (df['Age'] < 30),
    (df['Age'] >= 30) & (df['Age'] <= 50),
    (df['Age'] > 50)
]

# Define Age Group categories
age_groups = ['Young', 'Middle-aged', 'Senior']

# Add the new column 'Age Group' based on conditions
df['Age Group'] = np.select(conditions, age_groups, default='Unknown')

# Display the result
print(df[['Name', 'Age', 'Age Group']].head())
```

```python
#step22
# Identify columns with missing values
missing_values = df.isna().sum()

# Display columns with missing values
print(missing_values[missing_values > 0])
```

```python
#step23
# Fill missing values in Performance Score with the average score
df['Performance Score'] = df['Performance Score'].fillna(df['Performance Score'].mean())

# Display the result
print(df[['Name', 'Performance Score']].head())
```

```python
#step24
# Drop rows with any missing values
df = df.dropna()

# Display the result
print(df.head())
```

```python
#step25
# Replace missing values in Performance Score with the median by Department
df['Performance Score'] = df.groupby('Department')['Performance Score'].transform(lambda x: x.fillna(x.median()))

# Display the result
print(df[['Department', 'Performance Score']].head())
```

```python
#step26
# Calculate the average age of IT department employees
avg_age_it = df[df['Department'] == 'IT']['Age'].mean()

# Identify rows where the employee is in IT and Age is above the average
it_above_avg_age = df[(df['Department'] == 'IT') & (df['Age'] > avg_age_it)]

# Display the result
print(it_above_avg_age)
```

```python
#step27
# Remove rows with duplicate Name and Joining Date combinations
df = df.drop_duplicates(subset=['Name', 'Joining Date'])

# Display the result
print(df.head())
```

```python
#step28
# Remove rows where Experience is greater than Age - 18
df = df[df['Experience'] <= (df['Age'] - 18)]

# Display the result
print(df.head())
```

```python
#step29
# Sort the DataFrame by Salary in descending order
df_sorted_salary = df.sort_values(by='Salary', ascending=False)

# Display the result
print(df_sorted_salary.head())
```

```python
#step30
# Sort the DataFrame by Department and then by Joining Date in ascending order
df_sorted_dept_joining = df.sort_values(by=['Department', 'Joining Date'], ascending=[True, True])

# Display the result
print(df_sorted_dept_joining.head())
```

```python
#step31
# Rank employees by Salary within their Department
df['Salary Rank'] = df.groupby('Department')['Salary'].rank(ascending=False)

# Display the result
print(df[['Name', 'Department', 'Salary', 'Salary Rank']].head())
```

```python
#step32
# Sort by Experience in descending order, break ties with Performance Score in ascending order
df_sorted_exp_perf = df.sort_values(by=['Experience', 'Performance Score'], ascending=[False, True])

# Display the result
print(df_sorted_exp_perf.head())
```

```python
#step33
# Create a new DataFrame for HR employees with Salary above 5000
df_hr_high_salary = df[(df['Department'] == 'HR') & (df['Salary'] > 5000)]

# Display the result
print(df_hr_high_salary.head())
```

```python
#step34
# Filter out rows where the Location is "Chicago"
df_filtered_location = df[df['Location'] != 'Chicago']

# Display the result
print(df_filtered_location.head())
```

```python
#step35
# Delete rows where the Performance Score is below 2
df = df[df['Performance Score'] >= 2]

# Display the result
print(df.head())
```

```python
#step36
# Calculate the average Salary for each Department
avg_salary_by_dept = df.groupby('Department')['Salary'].mean()

# Display the result
print(avg_salary_by_dept)
```

```python
#step37
# Calculate the total Salary grouped by Status
total_salary_by_status = df.groupby('Status')['Salary'].sum()

# Display the result
print(total_salary_by_status)
```

```python
#step38
# Find the maximum Performance Score for each Location
max_perf_by_location = df.groupby('Location')['Performance Score'].max()

# Display the result
print(max_perf_by_location)
```

```python
#step39
# Extract the year from the 'Joining Date' column and create a new 'Joining Year' column
df['Joining Year'] = pd.to_datetime(df['Joining Date']).dt.year

# Calculate the total Salary paid to employees who joined in each year
total_salary_by_year = df.groupby('Joining Year')['Salary'].sum()

# Display the result
print(total_salary_by_year)
```

```python
#step40
# Find the department with the highest average Performance Score
avg_perf_by_dept = df.groupby('Department')['Performance Score'].mean()
top_dept = avg_perf_by_dept.idxmax()

# Display the result
print(f"The department with the highest average Performance Score is: {top_dept}")
```

```python
#step41
# Grouping by 'Location' and counting the number of employees
location_count = df.groupby('Location').size().reset_index(name='Employee Count')
print(location_count)
```

```python
#step42
salary_stats_by_status = df.groupby('Status')['Salary'].agg(['sum', 'mean', 'std']).reset_index()
print(salary_stats_by_status)
```

```python
#43
# Creating a new column 'Monthly Salary' by dividing 'Salary' by 12
df['Monthly Salary'] = df['Salary'].apply(lambda x: x / 12)

# Printing the 'ID', 'Salary', and 'Monthly Salary' columns
print(df[['ID', 'Salary', 'Monthly Salary']])
```

```python
#step44
df_uppercase = df.applymap(lambda x: x.upper() if isinstance(x, str) else x)
print(df_uppercase)
```

```python
#step45
budget_data = {
    'Department': ['HR', 'Finance'],
    'Budget': [500000, 600000]
}
df_budget = pd.DataFrame(budget_data)
df_merged = pd.merge(df, df_budget, on='Department', how='inner')
print(df_merged)
```

```python
#step46
# Correcting the merge operation by using 'ID' instead of 'Employee ID'
project_data = {
    'Employee ID': [1, 2, 3, 4, 5],
    'Recent Project': ['A', 'B', 'C', 'D', 'E']
}
df_projects = pd.DataFrame(project_data)

# Merge the dataframes on 'ID' from df and 'Employee ID' from df_projects
df_with_projects = pd.merge(df, df_projects, left_on='ID', right_on='Employee ID', how='inner')

# Print the merged dataframe
print(df_with_projects)
```

```python
#step47
team_lead_data = {
    'Department': ['HR', 'Finance'],
    'Team Lead': ['John', 'Jane']
}
df_team_lead = pd.DataFrame(team_lead_data)
df_with_team_lead = pd.merge(df, df_team_lead, on='Department', how='left')
print(df_with_team_lead)
```

```python
# Example of another DataFrame containing 'Project Count' for each 'ID'
project_data = {
    'ID': [1, 2, 3, 4, 5],
    'Project Count': [3, 2, 5, 1, 4]  # Example project count data
}
df_projects = pd.DataFrame(project_data)

# Merging the 'df' with the 'df_projects' DataFrame on 'ID'
df_with_project_count = pd.merge(df, df_projects, on='ID', how='inner')

# Calculate the average salary per project
df_with_project_count['Avg Salary per Project'] = df_with_project_count['Salary'] / df_with_project_count['Project Count']

# Print the updated DataFrame
print(df_with_project_count)
```

```python
# Calculate the average salary per department
avg_salary_by_department = df.groupby('Department')['Salary'].transform('mean')

# Calculate the salary increase percentage for each employee
df['Salary Increase Percentage'] = ((df['Salary'] - avg_salary_by_department) / avg_salary_by_department) * 100

# Print the updated DataFrame with 'ID', 'Salary', and 'Salary Increase Percentage'
print(df[['ID', 'Salary', 'Salary Increase Percentage']])
```
